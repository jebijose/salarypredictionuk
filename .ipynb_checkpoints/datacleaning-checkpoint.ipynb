```python
data = dt.loc[(dt["salary_offered"] != ' Salary not specified ') & (dt["salary_offered"] != ' Salary negotiable ')& (dt["salary_offered"] != ' Competitive salary ')& (dt["salary_offered"] != ' Commission only ')]
```

```python
data.info()
```

```python
# catagories field into annual, daily and hourly basis
data_annual_basis = data[data['salary_offered'].str.lower().str.contains("per annum", na =False)]
data_daily_basis = data[data['salary_offered'].str.lower().str.contains("per day", na =False)]
data_hourly_basis = data[data['salary_offered'].str.lower().str.contains("per hour", na =False)]
```

 Annual 

```python
data_annual_basis['salary_offered']
```

```python
#data_annual_basis[~data_annual_basis['salary_offered'].str.contains("£")]
annual_data = data_annual_basis[data_annual_basis['salary_offered'].str.contains("£")]
```

```python
annual_data['salary_offered']
```

```python
annual_data['salary_offered'].replace('[,]', '', regex = True, inplace = True)
```

```python
annual_data['salary'] = annual_data['salary_offered'].str.extractall('(\d+)').astype(int).groupby(level=0).mean()
```

```python
annual_data
```

Daily [multiply by 251 -average] UK 2022

```python
data_daily_basis['salary_offered']
```

```python
daily_data = data_daily_basis[data_daily_basis['salary_offered'].str.contains("£")]
```

```python
daily_data['salary_offered']
```

```python
daily_data['salary']  = daily_data['salary_offered'].str.extractall('£(\d+)').astype(int).groupby(level=0).mean()
```

daily_data['salary']

```python
daily_data['salary'] = daily_data['salary']*251
```

```python
daily_data
```

 hourly [ 1730 ] uk 2022

```python
data_hourly_basis['salary_offered']
```

```python
#data_hourly_basis[~data_hourly_basis['salary_offered'].str.contains("£")]
hourly_data = data_hourly_basis[data_hourly_basis['salary_offered'].str.contains("£")]
```

```python
#noise in the dataset
hourly_data = hourly_data[~hourly_data['salary_offered'].str.contains("22,000")]
hourly_data = hourly_data[~hourly_data['salary_offered'].str.contains("23,000")]
hourly_data = hourly_data[~hourly_data['salary_offered'].str.contains("30,000")]
```

```python
hourly_data['salary'] = hourly_data['salary_offered'].str.extractall('£(\d+.\d+)').astype(float).groupby(level=0).mean()
```

```python

hourly_data['salary'] = hourly_data['salary']*1730
```

```python
hourly_data
```
