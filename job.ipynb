```python
X = DATA.iloc[:,1:]
Y = DATA.iloc[:,0]
```

```python
# feature importances
from sklearn.ensemble import ExtraTreesRegressor
```

```python
model = ExtraTreesRegressor()
model.fit(X,Y)
```

```python
featureimportance = pd.Series(model.feature_importances_, index = X.columns)
featureimportance.nlargest(25).plot(kind='bar')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
n_estimators = [int(x) for x in np.linspace(start=100, stop=1200, num=12)]
#max features
max_features = ['auto','sqrt']
# depth of tree
max_depth = [int(x) for x in np.linspace(start=5, stop=30, num=6)]
# min sample split- node
min_samples_split = [2, 5, 10, 15, 100]
# min sample split- leaf
min_samples_leaf = [1, 2, 5, 10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
#create random grid
random_grid = {
    'n_estimators': n_estimators,
    'max_features' : max_features,
    'max_depth' : max_depth,
    'min_samples_split' : min_samples_split,
    'min_samples_leaf' : min_samples_leaf
}
print(random_grid)
```

```python
rf_random = RandomizedSearchCV(estimator=rf, param_distributions= random_grid, scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(x_train,y_train)
```

```python
rf_random.best_params_
```

```python
rf_random.best_score_
```

```python
predictions = rf_random.predict(x_test)
```

```python
predictions
```

```python
sns.distplot(y_test-predictions)
```

```python
plt.scatter(y_test,predictions)
```

```python
from sklearn import metrics
```

```python
print('mean absolute error:',metrics.mean_absolute_error(y_test,predictions))
print('mean squared error:',metrics.mean_squared_error(y_test,predictions))
print('root mean absolute error:', np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```
